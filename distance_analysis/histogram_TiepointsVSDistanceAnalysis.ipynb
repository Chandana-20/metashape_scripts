The histogram plot depicts the frequency distribution of tiepoints across different distance ranges. This visualization enables quantitative analysis of:

1. The spatial distribution of tiepoints relative to the camera center
2. Whether tiepoints fall within the depth of field range
3. Whether metashape selects feature points tightly within depth of field range.

```python
# Section 1: Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import json
import pandas as pd
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
```

```python
#loading data saved from code -> dist_from_tiepoints_to_camcenter.py

distances = np.load("distances_TwoMeters.npy")
tiepoints = np.load("tiepoints_TwoMeters.npy")
with open("tiepoint_distances_TwoMeters.json", "r") as f:
    tiepoint_distances = json.load(f)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(distances, kde=True, bins=50)
plt.title('Histogram of Distances from Camera Center to Tie Points')
plt.xlabel('Distance (m)')
plt.ylabel('Number of Tie Points')
plt.grid(True, alpha=0.3)
plt.savefig('distance_histogram.png', dpi=300, bbox_inches='tight')
plt.show()
```
